#### Problem Statement- From the given ‘Iris’ dataset, predict the optimum number of clusters and represent it visually.

Import required libraries in jupyter notebook

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

### Read the given csv file

```python
path="E:\Data Analysis\The_sparks_foundation\Task_3\iris.csv"
iris=pd.read_csv(path)
```

```python
iris.Species.unique()
```

Plot correlation matrix

```python
corrMatrix = iris.iloc[:,[1, 2, 3, 4]].corr()
sns.heatmap(corrMatrix, annot=True)
plt.show()
```

```python
X= iris.iloc[:, [1, 2, 3, 4]].values
```

### Optimum number of clusters

To get optimum no. of clusters use elbow method, start with 10 clusters to get optimum K

```python
wcss=[]

for k in range(1,11):
    kmeans=KMeans(n_clusters=k,init='k-means++',max_iter=300,random_state=0).fit(X)
    
    wcss.append(kmeans.inertia_)
    
len(wcss)
print(wcss)
```

Plot elbow graph using wcss- within cluster sum of squared distance

```python
plt.plot(wcss)
plt.title('Clusters vs WCSS')
plt.xlabel('Clusters')
plt.ylabel('WCSS')
plt.show()
```

### Apply Kmeans to the dataset

From above graph, choose optimum clusters, K=3, and predict clusters using Kmeans

```python
clusters=KMeans(n_clusters=3,init='k-means++',max_iter=300,random_state=0).fit_predict(X)
```

### Plot Graphs to represent columns by clusters

Plot scatter plot to represent clusters for first two columns, SepalLengthCm and SepalWidthCm

```python
plt.scatter(X[clusters == 0, 0], X[clusters == 0, 1], 
            s = 70, c = 'red', label = 'Iris-setosa')
plt.scatter(X[clusters == 1, 0], X[clusters == 1, 1], 
            s = 70, c = 'blue', label = 'Iris-versicolour')
plt.scatter(X[clusters == 2, 0], X[clusters == 2, 1],
            s = 70, c = 'green', label = 'Iris-virginica')

plt.legend()
```
